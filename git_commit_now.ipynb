# Commit Current Changes
This notebook stages all changes and creates a commit with a default message.

```python
import subprocess
from pathlib import Path
repo_path = Path('/app')
print('Working directory:', repo_path)
print('Current branch:')
print(subprocess.check_output(['git', 'branch', '--show-current'], cwd=repo_path, text=True).strip())
print('\nRemote origin:')
print(subprocess.check_output(['git', 'remote', '-v'], cwd=repo_path, text=True))
print('\nCurrent status:')
print(subprocess.check_output(['git', 'status', '--short'], cwd=repo_path, text=True))
print('\nSetting local git identity for this repo...')
subprocess.check_call(['git', 'config', 'user.email', 'noreply@github.com'], cwd=repo_path)
subprocess.check_call(['git', 'config', 'user.name', 'astacreative-bot'], cwd=repo_path)
print('Identity set to:')
print(subprocess.check_output(['git', 'config', 'user.name'], cwd=repo_path, text=True).strip())
print(subprocess.check_output(['git', 'config', 'user.email'], cwd=repo_path, text=True).strip())
print('\nStaging all changes...')
print(subprocess.check_output(['git', 'add', '--all'], cwd=repo_path, stderr=subprocess.STDOUT, text=True))
for excluded in ['backend/.env', 'frontend/.env']:
    try:
        subprocess.check_call(['git', 'reset', '--', excluded], cwd=repo_path)
        print(f'Excluded {excluded} from commit staging')
    except subprocess.CalledProcessError:
        pass
print('\nStaged changes:')
print(subprocess.check_output(['git', 'status', '--short'], cwd=repo_path, text=True))
commit_message = 'Commit latest code changes via automated notebook'
print('\nCommitting with message:', commit_message)
try:
    print(subprocess.check_output(['git', 'commit', '-m', commit_message], cwd=repo_path, stderr=subprocess.STDOUT, text=True))
except subprocess.CalledProcessError as exc:
    print('Commit command output:')
    print(exc.output)
print('\nPushing to origin...')
try:
    print(subprocess.check_output(['git', 'push', 'origin', 'HEAD'], cwd=repo_path, stderr=subprocess.STDOUT, text=True))
except subprocess.CalledProcessError as exc:
    print('Push failed:')
    print(exc.output)
```
