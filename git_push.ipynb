```python
# Verify Git Repository and Remote
Use shell commands to confirm the current directory is a git repository and show the configured GitHub remote.
</VSCode.Cell>
<VSCode.Cell language="python">
import os
from pathlib import Path
import subprocess
print('Current working directory:', Path.cwd())
print('Repository root exists:', (Path.cwd() / '.git').exists())
for cmd in [['git', 'status', '--short'], ['git', 'remote', '-v']]:
    print('\n$ ' + ' '.join(cmd))
    try:
        output = subprocess.check_output(cmd, stderr=subprocess.STDOUT, text=True)
        print(output)
    except subprocess.CalledProcessError as exc:
        print(exc.output)
</VSCode.Cell>
<VSCode.Cell language="markdown">
# Check Current Changes
Inspect the working tree for modified, deleted, or untracked files with git status and git diff.
</VSCode.Cell>
<VSCode.Cell language="python">
import subprocess
print('git status --short')
print(subprocess.check_output(['git', 'status', '--short'], text=True))
print('\n# git diff --stat')
print(subprocess.check_output(['git', 'diff', '--stat'], text=True))
</VSCode.Cell>
<VSCode.Cell language="markdown">
# Stage Files for Commit
Use git add to stage the latest changes and optionally review the staging area.
</VSCode.Cell>
<VSCode.Cell language="python">
import subprocess
print('Staging all changes...')
print(subprocess.check_output(['git', 'add', '--all'], text=True))
print('\n# Staged changes (git status --short)')
print(subprocess.check_output(['git', 'status', '--short'], text=True))
</VSCode.Cell>
<VSCode.Cell language="markdown">
# Create a Commit with a Message
Commit the staged changes with a descriptive message using git commit.
</VSCode.Cell>
<VSCode.Cell language="python">
import subprocess
commit_message = 'Update latest changes via notebook automation'
print('Committing with message:', commit_message)
try:
    print(subprocess.check_output(['git', 'commit', '-m', commit_message], stderr=subprocess.STDOUT, text=True))
except subprocess.CalledProcessError as exc:
    print('Commit command output:')
    print(exc.output)
</VSCode.Cell>
<VSCode.Cell language="markdown">
# Push the Latest Commit to GitHub
Push the new commit to the remote GitHub branch with git push.
</VSCode.Cell>
<VSCode.Cell language="python">
import subprocess
print('Pushing to origin...')
try:
    print(subprocess.check_output(['git', 'push'], stderr=subprocess.STDOUT, text=True))
except subprocess.CalledProcessError as exc:
    print('Push failed:')
    print(exc.output)
</VSCode.Cell>
<VSCode.Cell language="markdown">
# Confirm Push and Show Remote Status
Verify the push succeeded and display the remote branch status with git log and git remote show.
</VSCode.Cell>
<VSCode.Cell language="python">
import subprocess
print('\n# Last 5 commits')
print(subprocess.check_output(['git', 'log', '--oneline', '-5'], text=True))
print('\n# Remote show origin')
print(subprocess.check_output(['git', 'remote', 'show', 'origin'], text=True))
```
